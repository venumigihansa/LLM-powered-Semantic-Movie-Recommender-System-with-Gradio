```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows")

print("Path to dataset files:", path)
```

```python
import pandas as pd
movies=pd.read_csv(f"{path}/imdb_top_1000.csv")
```

```python
movies
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax=plt.axes()
sns.heatmap(movies.isna().transpose(),cbar=False, ax=ax)
plt.xlabel("Columns")
plt.ylabel("missing values")
plt.show()
```

```python
movies["Genre"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
movies["words_in_overview"]= movies["Overview"].str.split().str.len()
```

```python
movies
```

```python
plt.figure(figsize=(10,6))
sns.histplot(movies["words_in_overview"], bins=30, kde=False)
plt.title("Distribution of Words in Overview")
plt.xlabel("Number of Words in Overview")
plt.ylabel("Number of Movies")
plt.show()
```

```python
movies.loc[movies["words_in_overview"].between(1,5),"Overview"]
```

```python
movies.loc[movies["words_in_overview"].between(5,15),"Overview"]
```

```python
movies.loc[movies["words_in_overview"].between(15,25),"Overview"]
```

```python
movies["tagged_description"] = movies.index.astype(str) + " " + movies["Overview"].astype(str)
```

```python
movies
```

```python
(
    movies
    .drop(["words_in_overview"], axis=1)
    .to_csv("movies_cleaned.csv", index = False)
)
```
